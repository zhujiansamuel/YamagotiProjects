```python
df = pd.read_csv(filepath)
df = df.drop(['Unnamed: 0', 'Unnamed: 2'], axis=1, errors='ignore')
df_1 = pd.read_csv(filepath_1)
df_2 = pd.read_csv(filepath_2)
df_3 = pd.read_csv(filepath_3)
df_4 = pd.read_csv(filepath_4)
df_5 = pd.read_csv(filepath_5)
df_6 = pd.read_csv(filepath_6)
df_7 = pd.read_csv(filepath_7)
df_8 = pd.read_csv(filepath_8)
df_9 = pd.read_csv(filepath_9)
df_10 = pd.read_csv(filepath_10)
df_11 = pd.read_csv(filepath_11)
df_12 = pd.read_csv(filepath_12)
```

```python
columns_name = ["第１回","第2回","第３回","第４回","第５回","第6回","第7&8回","第9回","第10回","第11&12回","第13回","第14回"]
df_list = [df_1,df_2,df_3,df_4,df_5,df_6,df_7,df_8,df_9,df_10,df_11,df_12]
for index, df_s in enumerate(df_list):
  for row_1 in df_s.itertuples():
    insert = True
    for row in df.itertuples():
      # print(getattr(row_1,"学生番号"),"------",getattr(row, "学籍番号"))
      # print(type(getattr(row_1,"学生番号")))
      # print(type(getattr(row, "学籍番号")))
      if not pd.isna(getattr(row_1,"学生番号")):
        if getattr(row_1,"学生番号") == str(getattr(row, "学籍番号")):
          print(getattr(row_1,"学生番号"),"------",getattr(row, "学籍番号"))
          df.at[row.Index,columns_name[index]] = 1
          insert = False
    if insert:
      print(getattr(row_1,"学生番号"))
      df = pd.concat([df, pd.DataFrame({'学籍番号':[getattr(row_1,"学生番号")],columns_name[index]:1})], ignore_index=True)
```

```python
def core_so2py():
    so_file = open('.core.so', 'rb')
    buffer = so_file.read(2)
    r, = struct.unpack('h', buffer)
    buffer = so_file.read(r)
    pri = buffer
    strs = bytearray()
    while True:
        temp = so_file.read(4)
        if len(temp) == 0:
            so_file.close()
            break
        size, = struct.unpack('i', temp)
        d = so_file.read(size)
        privkey = rsa.PrivateKey.load_pkcs1(pri)
        strs.extend(rsa.decrypt(d, privkey))
    print(strs)
    with open("core.py", 'wb') as f:
        f.write(strs)
        
core_so2py()
```
